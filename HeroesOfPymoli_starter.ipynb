```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file= "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file)
purchase_dropdupes_df = purchase_data_df.drop_duplicates("SN", inplace=False)
purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
#calculate total players and insert into new column
TotalPlayers = len(purchase_data_df["SN"].unique())
purchase_data_df["Total Players"] = TotalPlayers
#find and display just the total players
Players = purchase_data_df[["Total Players"]]
Players.head(1)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#calculate new values
UniqueItems = len(purchase_data_df["Item Name"].unique())
#average price
AveragePrice = round((purchase_data_df["Price"].mean()),2)
#count number of purchases
TotalPurchases = purchase_data_df["Purchase ID"].count()
#sum revenue
Revenue = purchase_data_df["Price"].sum()


#insert into dictionary
analysis_dict = [{"Number of Unique Items":UniqueItems, "Average Price":AveragePrice, "Number of Purchases": TotalPurchases, "Total Revenue": Revenue} ]
purchasing_analysis_df = pd.DataFrame(analysis_dict)

#format Price and Revenue field
purchasing_analysis_df["Average Price"] = purchasing_analysis_df["Average Price"].map("${:,}".format)
purchasing_analysis_df["Total Revenue"] = purchasing_analysis_df["Total Revenue"].map("${:,}".format)
#reorder columns
purchasing_analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#group by gender - excluding duplicate SN 
gendergroups = purchase_dropdupes_df.groupby(['Gender'])
#calculate unique number of players of each gender
gendergroupscount = gendergroups['SN'].count() #picked 'SN' as its a unique value to iterate through
#calculate percent gender/ total players
genderpercent = gendergroupscount/ TotalPlayers *100

genderpercent_df = pd.DataFrame({"Player Count": gendergroupscount, "Percent": genderpercent })
genderpercent_df["Player Count"] = genderpercent_df["Player Count"].map("{:,}".format)
genderpercent_df["Percent"] = genderpercent_df["Percent"].map("{:.2f}%".format)
genderpercent_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#group data by gender including duplicate purchases
purchasegendergroups = purchase_data_df.groupby(['Gender'])
#number of purchases made by each gender
purchase_count = purchasegendergroups['Purchase ID'].sum()
#Average price by gender
average_price =purchasegendergroups['Price'].mean()
#Total Purchase Value
totalpurchasevalue = purchasegendergroups['Price'].sum()
#purchase total per person
average_purchase_total_per_person = average_price / TotalPlayers *100

#establish dictionary/ datafram of new values
genderpurchaseanalysis_df = pd.DataFrame({"Purchase Count":purchase_count, "Average Purchase Price":average_price, "Total Purchase Value": totalpurchasevalue, "Avg Total Purchase per Person":average_purchase_total_per_person})
#format dataframe
genderpurchaseanalysis_df["Average Purchase Price"] = genderpurchaseanalysis_df["Average Purchase Price"].map("${:,.2f}".format)
genderpurchaseanalysis_df["Total Purchase Value"] = genderpurchaseanalysis_df["Total Purchase Value"].map("${:,.2f}".format)
genderpurchaseanalysis_df["Avg Total Purchase per Person"] = genderpurchaseanalysis_df["Avg Total Purchase per Person"].map("${:,.2f}".format)
#display dataframe
genderpurchaseanalysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#create bins for age ranges
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9,34.9, 39.9, 99.9 ]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#create Age Category Column
purchase_dropdupes_df["Age Category"]= pd.cut(purchase_dropdupes_df["Age"], bins, labels=group_names, include_lowest=True )

#group players by Age Category
age_group = purchase_dropdupes_df.groupby("Age Category")
#count of players in each age range
age_count = age_group['SN'].count() #picked 'SN' as a unique value
#calculate percentage of players in each age category
age_percent = age_count/TotalPlayers *100
 
#set new datarframe
age_group_df = pd.DataFrame({"Total Count": age_count, "Percentage of Players": age_percent})
age_group_df["Percentage of Players"] = age_group_df["Percentage of Players"].map("{:.2f}%".format)
age_group_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#create Age Category Column based on full data set for all purchases
purchase_data_df["Age Category"]= pd.cut(purchase_data_df["Age"], bins, labels=group_names, include_lowest=True )

#group players by Age Category
priceage_group = purchase_data_df.groupby("Age Category")
#count of players in each age range
priceage_count = priceage_group['SN'].count() #picked 'SN' as a unique value
#calculate average purchase price
priceage_average = priceage_group['Price'].mean()
#purchase total
priceage_sum = priceage_group['Price'].sum()
#calculate percentage of players in each age category
priceage_percent = priceage_sum/age_count

#set new datarframe
price_age_group_df = pd.DataFrame({"Purchase Count": priceage_count, "Average Purchase Price": priceage_average, "Total Purchase Value": priceage_sum,  "Average Total Purchase per Person":priceage_percent })
price_age_group_df["Average Total Purchase per Person"] = price_age_group_df["Average Total Purchase per Person"].map("${:.2f}".format)
price_age_group_df["Average Purchase Price"] = price_age_group_df["Average Purchase Price"].map("${:.2f}".format)
price_age_group_df["Total Purchase Value"] = price_age_group_df["Total Purchase Value"].map("${:.2f}".format)
price_age_group_df

```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#groupby SN for Number of purchases
top_spenders = purchase_data_df.groupby('SN')
#how many purchases per user
purchase_count = top_spenders['Purchase ID'].count()
#average purchase price
purchase_avg_price = top_spenders['Price'].mean()
#total purchase value
purchase_total_amount = top_spenders['Price'].sum()

#create and format dateframe
top_spenders_df = pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price":purchase_avg_price, "Total Purchase Value":purchase_total_amount})
top_spenders_df["Total Purchase Value"] = top_spenders_df["Total Purchase Value"].map("${:.2f}".format)
top_spenders_df["Average Purchase Price"] = top_spenders_df["Average Purchase Price"].map("${:.2f}".format)
top_spenders_df.head()
```
